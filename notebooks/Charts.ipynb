```python
import matplotlib.pyplot as plt

# Datos de ejemplo
idiomas = ['Ingls', 'Espaol', 'Francs', 'Italiano', 'Cataln', 'Euskera', 'Blgaro', 'Armenio', 'Georgiano', 'Uighur']
instancias_entrenamiento = [100000000, 20000000, 9000000, 2000000, 750000, 175000, 70000, 7500, 1000, 100]

accuracy_1000inst = [0.97, 0.70, 0.78, 0.74, 0.70, 0.54, 0.69, 0.59, 0.65, 0.49]
accuracy_250inst = [0.98, 0.71, 0.78, 0.77, 0.73, 0.59, 0.72, 0.64, 0.70, 0.54]

# Crear la figura y los ejes
fig, ax1 = plt.subplots()

# Crear la grfica de barras para el nmero de instancias de entrenamiento por idioma
bars = ax1.bar(idiomas, instancias_entrenamiento, color='#1D7CCB', label='Instancias de entrenamiento')
ax1.set_ylabel('Nmero de instancias de entrenamiento', color='black')
ax1.tick_params(axis='y', labelcolor='#1D7CCB')

ax1.set_yscale('log')

# Rotar los nombres de los idiomas en el eje x
plt.xticks(rotation=90)  # Rotar 90 grados

# Crear la grfica de lnea para el nmero de instancias correctamente clasificadas (accuracy)
ax2 = ax1.twinx()  # Crear un segundo eje y
ax2.plot(idiomas, accuracy_1000inst, color='red', marker='o', label='Accuracy 1000 inst')
ax2.plot(idiomas, accuracy_250inst, color='blue', marker='s', label='Accuracy 250 inst')  # Nueva lnea de accuracy
ax2.set_ylabel('Accuracy del clasificador', color='black')
ax2.tick_params(axis='y')

# Aadir leyendas
plt.legend()

# Mostrar el grfico
plt.title('Relacin entre el nmero de instancias de entrenamiento y aciertos por idioma\n\n\n')
plt.tight_layout()  # Ajustar diseo para evitar superposicin
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

# Datos de ejemplo
idiomas = ['Ingls', 'Espaol', 'Francs', 'Italiano', 'Cataln', 'Euskera', 'Blgaro', 'Armenio', 'Georgiano', 'Uighur']
non_suicide = [0.97, 0.76, 0.81, 0.79, 0.76, 0.68, 0.76, 0.70, 0.73, 0.65]
suicide = [0.97, 0.59, 0.73, 0.66, 0.60, 0.20, 0.57, 0.33, 0.49, 0.01]

# Crear la figura y los ejes
fig, ax1 = plt.subplots()

ax1.plot(idiomas, non_suicide, color='red', marker='o', label='Non-suicide')
ax1.plot(idiomas, suicide, color='blue', marker='s', label='Suicide')

ax1.set_ylabel('F-score de la inferencia', color='black')

ax1.tick_params(axis='y')

ax1.set_yscale('log')

# Usar ScalarFormatter para mostrar los valores directos en lugar de notacin cientfica en el eje y
ax1.yaxis.set_major_formatter(ScalarFormatter())

# Rotar los nombres de los idiomas en el eje x
plt.xticks(rotation=90)

# Aadir leyendas
plt.legend()

# Mostrar el grfico
plt.title('Relacin entre el nmero de instancias de entrenamiento y aciertos por idioma\n\n\n')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos proporcionados
f1_score = 0.98  # F1-score promediado sin alterar el orden de las palabras
accuracy = 0.985  # Accuracy sin alterar el orden de las palabras

f1_score_alterado = 0.97  # F1-score promediado con el orden de las palabras alterado
accuracy_alterado = 0.973  # Accuracy con el orden de las palabras alterado

labels = ['F-Score', 'Accuracy']
values = [f1_score, accuracy]
values_alterado = [f1_score_alterado, accuracy_alterado]

x = range(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x, values, width, label='Orden por defecto', color='#1D7CCB')  # Azul
rects2 = ax.bar([i + width for i in x], values_alterado, width, label='Orden alterado', color='red')  # Rojo

# Etiquetas, ttulo y leyenda
ax.set_xlabel('Mtricas')
ax.set_ylabel('Scores')
ax.set_title('Diferencias en Mtricas (Weighted Average)')
ax.set_xticks([i + width/2 for i in x])
ax.set_xticklabels(labels)
ax.legend()

# Mostrar las etiquetas sobre las barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # Desplazamiento vertical de la etiqueta
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

# Ampliar an ms el rango del eje y para evitar superposicin con la leyenda
ax.set_ylim(0, 1.25)  # Rango de valores ajustado

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
idiomas = ['Ingls', 'Espaol', 'Francs', 'Italiano', 'Cataln', 'Euskera', 'Blgaro', 'Armenio', 'Georgiano', 'Uighur']
instancias_entrenamiento = [100000000, 20000000, 9000000, 2000000, 750000, 175000, 70000, 7500, 1000, 100]

accuracy_1000inst = [0.97, 0.70, 0.78, 0.74, 0.70, 0.54, 0.69, 0.59, 0.65, 0.49]
accuracy_1000inst_random = [0.96, 0.84, 0.88, 0.82, 0.81, 0.54, 0.76, 0.60, 0.66, 0.49]

# Crear la figura y los ejes
fig, ax1 = plt.subplots()

# Crear la grfica de barras para el nmero de instancias de entrenamiento por idioma
bars = ax1.bar(idiomas, instancias_entrenamiento, color='#1D7CCB', label='Instancias de entrenamiento')
ax1.set_ylabel('Nmero de instancias de entrenamiento', color='black')
ax1.tick_params(axis='y', labelcolor='#1D7CCB')

ax1.set_yscale('log')

# Rotar los nombres de los idiomas en el eje x
plt.xticks(rotation=90)  # Rotar 90 grados

# Crear la grfica de lnea para el nmero de instancias correctamente clasificadas (accuracy)
ax2 = ax1.twinx()  # Crear un segundo eje y
ax2.plot(idiomas, accuracy_1000inst, color='blue', marker='o', label='Accuracy 1000 inst')
ax2.plot(idiomas, accuracy_1000inst_random, color='red', marker='s', label='Accuracy 1000 inst (randomized)')
ax2.set_ylabel('Accuracy del clasificador', color='black')
ax2.tick_params(axis='y')

# Aadir leyendas
plt.legend()

# Mostrar el grfico
plt.title('Relacin entre el nmero de instancias de entrenamiento y aciertos por idioma\n\n\n')
plt.tight_layout()  # Ajustar diseo para evitar superposicin
plt.show()
```
