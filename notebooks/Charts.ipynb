```python
import matplotlib.pyplot as plt

# Datos de ejemplo
idiomas = ['Inglés', 'Español', 'Francés', 'Italiano', 'Catalan', 'Euskera', 'Bulgaro', 'Armenio', 'Georgiano', 'Uighur']
instancias_entrenamiento = [100000000, 20000000, 9000000, 2000000, 750000, 175000, 70000, 7500, 1000, 100]
accuracy_500inst = [0.97, 0.70, 0.77, 0.74, 0.7, 0.53, 0.69, 0.58, 0.65, 0.47]

# Nuevos datos de accuracy para diferentes casos
accuracy_1000inst = [0.95, 0.72, 0.79, 0.76, 0.68, 0.56, 0.71, 0.62, 0.67, 0.51]
accuracy_2000inst = [0.92, 0.75, 0.81, 0.78, 0.65, 0.60, 0.74, 0.67, 0.71, 0.58]

# Crear la figura y los ejes
fig, ax1 = plt.subplots()

# Crear la gráfica de barras para el número de instancias de entrenamiento por idioma
bars = ax1.bar(idiomas, instancias_entrenamiento, color='#1D7CCB', label='Instancias de entrenamiento')
ax1.set_ylabel('Número de instancias de entrenamiento', color='black')
ax1.tick_params(axis='y', labelcolor='#1D7CCB')

ax1.set_yscale('log')

# Rotar los nombres de los idiomas en el eje x
plt.xticks(rotation=90)  # Rotar 90 grados

# Crear la gráfica de línea para el número de instancias correctamente clasificadas (accuracy)
ax2 = ax1.twinx()  # Crear un segundo eje y
ax2.plot(idiomas, accuracy_500inst, color='red', marker='o', label='Accuracy 500 inst')
ax2.plot(idiomas, accuracy_1000inst, color='green', marker='x', label='Accuracy 1000 inst')  # Nueva línea de accuracy
ax2.plot(idiomas, accuracy_2000inst, color='blue', marker='s', label='Accuracy 2000 inst')  # Otra nueva línea de accuracy
ax2.set_ylabel('Accuracy del clasificador', color='black')
ax2.tick_params(axis='y', labelcolor='red')

# Añadir leyendas
plt.legend()

# Mostrar el gráfico
plt.title('Relación entre el número de instancias de entrenamiento y aciertos por idioma\n\n\n')
plt.tight_layout()  # Ajustar diseño para evitar superposición
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

# Datos de ejemplo
idiomas = ['Inglés', 'Español', 'Francés', 'Italiano', 'Catalan', 'Euskera', 'Bulgaro', 'Armenio', 'Georgiano', 'Uighur']
non_suicide = [0.97, 0.75, 0.80, 0.78, 0.76, 0.66, 0.75, 0.69, 0.72, 0.64]
suicide = [0.97, 0.60, 0.73, 0.68, 0.62, 0.20, 0.59, 0.35, 0.51, 0.01]

# Crear la figura y los ejes
fig, ax1 = plt.subplots()

ax1.plot(idiomas, non_suicide, color='red', marker='o', label='Non-suicide')
ax1.plot(idiomas, suicide, color='blue', marker='s', label='Suicide')

ax1.set_ylabel('F-score de la inferencia', color='black')

ax1.tick_params(axis='y', labelcolor='red')

ax1.set_yscale('log')

# Usar ScalarFormatter para mostrar los valores directos en lugar de notación científica en el eje y
ax1.yaxis.set_major_formatter(ScalarFormatter())

# Rotar los nombres de los idiomas en el eje x
plt.xticks(rotation=90)

# Añadir leyendas
plt.legend()

# Mostrar el gráfico
plt.title('Relación entre el número de instancias de entrenamiento y aciertos por idioma\n\n\n')
plt.tight_layout()
plt.show()
```
